```python
# Read the CSV file and create a count of people by illness
import pandas as pd

df = pd.read_csv('DoctorVisits-DA .csv')

# Group by illness and count the number of people
illness_counts = df.groupby('illness')['visits'].count().reset_index()
illness_counts.columns = ['Illness Count', 'Number of People']

# Sort by illness count
illness_counts = illness_counts.sort_values('Illness Count')

print("Distribution of People by Illness Count:")
print(illness_counts)

# Create a bar plot to visualize the distribution
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.barplot(data=illness_counts, x='Illness Count', y='Number of People')
plt.title('Distribution of People by Illness Count')
plt.xlabel('Number of Illnesses')
plt.ylabel('Number of People')
plt.show()
```

```python
# Create a pie chart to visualize the distribution of people by illness count
plt.figure(figsize=(8, 8))
plt.pie(illness_counts['Number of People'], labels=illness_counts['Illness Count'], 
        autopct='%1.1f%%', startangle=140)
plt.title('Distribution of People by Illness Count')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
Based on the pie chart analysis:

30.0% of people had 1 illness, making it the most common category
29.9% had no illnesses (0)
18.2% experienced 2 illnesses
10.4% had 3 illnesses
5.3% had 4 illnesses
4.5% had 5 illnesses
This shows that approximately 60% of the population had 0-1 illnesses, while a smaller percentage experienced multiple illnesses during the observed period
.
```

```python
import seaborn as sns

# Plotting the bar chart
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='gender', y='reduced', ci=None, estimator=sum)
plt.title('Total Reduced Days of Activity Due to Illness by Gender')
plt.xlabel('Gender')
plt.ylabel('Total Reduced Days')
plt.show()
```

```python
# Calculate total reduced days for each gender
gender_reduced_days = df.groupby('gender')['reduced'].sum()

gender_labels = gender_reduced_days.index
sizes = gender_reduced_days.values

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=gender_labels, autopct='%1.1f%%', startangle=90)
plt.title('Percentage of Reduced Days of Activity Due to Illness by Gender')
plt.axis('equal')
plt.show()
```

```python
Based on both visualizations:

Bar Chart and Pie Chart Analysis:

Females account for 60.9% of total reduced activity days
Males account for 39.1% of total reduced activity days
This indicates that females in the dataset reported more days of reduced activity due to illness compared to males, suggesting either:

Higher illness frequency among females
More significant impact of illnesses on daily activities for females
Potentially different reporting patterns between genders
```

```python
Relation between Health status and Activity reduction using heat map chart.
```

```python
import seaborn as sns
import numpy as np

# Create a cross-tabulation of health status and reduced activity
heatmap_data = pd.crosstab(df['health'], df['reduced'])

# Create heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, annot=True, fmt='d', cmap='YlOrRd')
plt.title('Relationship between Health Status and Activity Reduction')
plt.xlabel('Days of Reduced Activity')
plt.ylabel('Health Status')
plt.show()

# Calculate correlation
correlation = df['health'].astype('int64').corr(df['reduced'])
print("\
Correlation between Health Status and Reduced Activity:", round(correlation, 3))
```

```python
The heatmap shows the relationship between health status (1-excellent to 4-poor) and days of reduced activity. There's a positive correlation of 0.28, indicating that poorer health status tends to be associated with more days of reduced activity.
```

```python
import seaborn as sns
import numpy as np

# Calculate income categories
df['income_category'] = pd.qcut(df['income'], q=3, labels=['Low', 'Medium', 'High'])

# Bar Chart
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='income_category')
plt.title('Distribution of Income Categories')
plt.xlabel('Income Level')
plt.ylabel('Count')
plt.show()

# Pie Chart
income_dist = df['income_category'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(income_dist.values, labels=income_dist.index, autopct='%1.1f%%')
plt.title('Income Distribution')
plt.show()

# Heat Map
income_health = pd.crosstab(df['income_category'], df['health'])
plt.figure(figsize=(10, 6))
sns.heatmap(income_health, annot=True, fmt='d', cmap='YlOrRd')
plt.title('Income vs Health Status')
plt.show()

print("\
Income Statistics:")
print("Maximum Income:", df['income'].max())
print("Minimum Income:", df['income'].min())
print("Median Income:", df['income'].median())
```

```python
Key findings:

Income distribution is fairly balanced across low, medium, and high categories (approximately 33% each)
The heatmap reveals that higher income groups tend to report better health status
Income range in the dataset: Min = 0.0, Max = 1.5, Median = 0.55
```
